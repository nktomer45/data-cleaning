```python

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
y=sc.fit_transform(y.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,test_x,train_y,test_y=train_test_split(data,y,test_size=.2,random_state=True)
```

```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.metrics import make_scorer,r2_score
```

```python
def test_model(model,x=x_train,y=train_y):
    cv=KFold(n_splits=3,shuffle=True,random_state=45)
    r2=make_scorer(r2_score)
    r2_val_score=cross_val_score(model,x,y,cv=cv,scoring=r2)
    score=[r2_val_score.mean()]
    return score
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
test_model(lr)
```

```python
from sklearn.svm import SVR
svr_reg=SVR(kernel='rbf',degree=4,gamma=.2)
test_model(svr_reg)
```

```python
svr_reg.fit(x_train,train_y)
```

```python
y_pred=svr_reg.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=accuracy_score(test_y,y_pred)*100
cm
```
